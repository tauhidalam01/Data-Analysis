```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv('sales_data.csv')

# Set Theme
sns.set_theme(style="whitegrid")

# Create Dashboard Layout
fig, axs = plt.subplots(2, 2, figsize=(14,10))

# 1. Line Chart: Monthly Sales Trend
sns.lineplot(x='Month', y='Sales', data=df, ax=axs[0,0], marker='o', color='blue')
axs[0,0].set_title('Monthly Sales Trend', fontsize=14)
axs[0,0].set_xlabel('Month')
axs[0,0].set_ylabel('Sales')

# 2. Bar Chart: Monthly Profit
sns.barplot(x='Month', y='Profit', data=df, ax=axs[0,1], palette='viridis')
axs[0,1].set_title('Monthly Profit', fontsize=14)
axs[0,1].set_xlabel('Month')
axs[0,1].set_ylabel('Profit')

# 3. Scatter Plot: Sales vs Profit
sns.scatterplot(x='Sales', y='Profit', data=df, ax=axs[1,0], s=100, color='green')
axs[1,0].set_title('Sales vs Profit', fontsize=14)
axs[1,0].set_xlabel('Sales')
axs[1,0].set_ylabel('Profit')

# 4. Pie Chart: Sales Distribution
axs[1,1].pie(df['Sales'], labels=df['Month'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
axs[1,1].set_title('Sales Distribution', fontsize=14)

plt.tight_layout()
plt.show()
```
